```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.goibibo.com/flights/air-BOM-BLR-20241015--1-0-0-E-D?'
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
```

```python
response = requests.get(url, headers=headers)
```

```python
response.raise_for_status()
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
flights = soup.find_all('span', class_='srp-card-uistyles__Time-sc-3flq99-15')
flights
```

```python
airlines = []
departure_times = []
arrival_times = []
durations = []
prices = []
```

```python
for flight in flights:
    try:
        airline = flight.find('span', class_='SRP_Carrier_Name').text.strip()
        departure_time = flight.find('div', class_='departureTime').text.strip()
        arrival_time = flight.find('div', class_='srp-card-uistyles__Time-sc-3flq99-15 iHpsco.padT10.f600').text.strip()
        duration = flight.find('div', class_='duration').text.strip()
        price = flight.find('div', class_='price').text.strip()

        airlines.append(airline)
        departure_times.append(departure_time)
        arrival_times.append(arrival_time)
        durations.append(duration)
        prices.append(price)
    except AttributeError:
        continue
```

```python
data = {
    'Airline': airlines,
    'Arrival Time': arrival_times,
}
```

```python
df = pd.DataFrame(data)
```
